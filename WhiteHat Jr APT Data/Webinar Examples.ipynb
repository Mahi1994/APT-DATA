### Bill calculation


You went to a restaurant and ordered a few items. Now that food has been ordered you want to predict what the the total bill would be using Python programming knowledge you have gained.

#### Instructions

1.The price of each food item(in dollars) is given

2.Calculate the total bill by adding the prices and store the result in a variable called 'total_bill'

3.Tax is calculated as 5% of the total bill. Calculate it and store it in a variable called 'tax'

4.Add tax with total bill and store the result in 'final_bill'.

```
# Price of food item 1
item_1 = 12

# Price of food item 2
item_2 = 5.5

# Price of food item 3
item_3 = 7

# Price of food item 4
item_4 = 6

# Calculate the total bill
total_bill=item_1+item_2+item_3+item_4

# Calculate the tax
tax=(5/100)*total_bill

# Calculate the final bill
final_bill=total_bill+tax


# Print the final bill
print(final_bill)
```

### Best cellular plan

Find the most economical cellular plan in terms of cheapest per minute calling cost

```
# Details of Plan 1
nom_1 = 100
price_1 = 35

# Calculate the value of plan 1
cpm_1=price_1/nom_1


# Details of Plan 2
nom_2 = 125
price_2 = 39

# Calculate the value of plan 2
cpm_2=price_2/nom_2


# Details of Plan 3
nom_3 = 180
price_3 = 45

# Calculate the value of plan 3
cpm_3=price_3/nom_3


print(cpm_1,cpm_2,cpm_3)
# Find minumim of three
economical_rate=min(cpm_1,cpm_2,cpm_3)
print(economical_rate)
```

# Speed of NumPy vs Python lists
Code snippet Speed comparison while performing element-wise addition of two Python lists and element-wise addition of two NumPy.

```
# import packages
import time
import numpy as np

# initialize variable
num = 10000

# initialize lists
l1, l2 = [i for i in range(num)], [i+2 for i in range(num)]

# initialize arrays
a1, a2 = np.array(l1), np.array(l2)

# start time
start_list = time.time()

# element-wise addition for both lists
sum_lists = [i+j for i, j in zip(l1, l2)]

# stop time
stop_list = time.time()

# display time
print(stop_list - start_list)


# start time
start_array = time.time()

# element-wise addition for both arrays
sum_arrays = a1 + a2

# stop time
stop_array = time.time()

# display time
print(stop_array - start_array)
```
